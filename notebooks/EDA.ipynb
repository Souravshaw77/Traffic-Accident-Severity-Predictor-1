```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
pd.set_option("display.max_columns", None)
```

```python
RAW_PATH = "../data/raw/road_accidents_india.csv"

df = pd.read_csv(RAW_PATH)
df.head()
```

```python
df.shape, df.dtypes
```

```python
df.describe(include="all").transpose()
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()

df.isnull().sum()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df, x="Accident_severity", palette="viridis")
plt.title("Target Class Distribution")
plt.xticks(rotation=45)
plt.show()

df["Accident_severity"].value_counts()
```

```python
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()

print("Categorical columns:", categorical_cols)

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, palette="coolwarm")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

df[numeric_cols].hist(figsize=(12,10), bins=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Numeric Feature Correlation Heatmap")
plt.show()
```

```python
key_numeric = ["Number_of_vehicles_involved", "Number_of_casualties"]

for col in key_numeric:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=df, x="Accident_severity", y=col)
    plt.title(f"{col} by Accident Severity")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# If 'Time' exists, convert HH:MM to hour
if "Time" in df.columns:
    df["Hour"] = df["Time"].str.split(":").str[0].astype(int)
    sns.countplot(data=df, x="Hour")
    plt.title("Accident Hour Distribution")
    plt.show()
```

```python
eda_summary = {
    "rows": df.shape[0],
    "cols": df.shape[1],
    "class_distribution": df["Accident_severity"].value_counts().to_dict(),
    "numeric_cols": numeric_cols,
    "categorical_cols": categorical_cols
}

eda_summary
```
